## Slicing Summary

`x[start:end:step] = [x[start], x[start+step], x[start+2*step], ... x[end-1]]`

If `start` not specified, assumes `0`

If `end` not specified, assumes `len(x)`

If `step` not specified, assumes `1`

**Boolean indexing**: `x[ind]` where `ind` is an array of `True` and `False` with the same size as `x` 

```python
x = np.array([4, 1, -1, 2, 4, 5])
print(x[1:4]) # [1, -1, 2] 
print(x[1:]) # [1, -1, 2, 4, 5]
print(x[:4]) # []
print(x[:]) # []
print(x[2::2]) # []
print(x[:6:2]) # []
```

## Example 2

```python
I = io.imread("/content/images.jpg")
```

```python
I.shape
```

```python
height, width, nchannels = I.shape
print("Height =", height)
print("Width =", width)
print("Number of channels =", nchannels)
```

```python
io.imshow(I)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Red channel
axes[0].imshow(I[:, :, 0], cmap="gray")
axes[0].axis("off")
axes[0].set_title("Red")

# Green channel
axes[1].imshow(I[:, :, 1], cmap="gray")
axes[1].axis("off")
axes[1].set_title("Green")

# Blue channel
axes[2].imshow(I[:, :, 2], cmap="gray")
axes[2].axis("off")
axes[2].set_title("Blue")
```

```python
# Only the left half of the image
I_cropped = I[:, :width//2, :]
io.imshow(I_cropped)
plt.axis("off")
```

```python
# Only the  bottom half of the image
I_cropped = I[height//2:, :, :]
io.imshow(I_cropped)
plt.axis("off")
```

```python
# flip along widht (with respect to y axis)
I_flipped = I[:, ::-1, :]
io.imshow(I_flipped)
plt.axis("off")
```

```python
# flip along height (with respect to x axis)
I_flipped = I[::-1, :, :]
io.imshow(I_flipped)
plt.axis("off")
```

```python
# Rotate 180 degree
I_rotated = I[::-1, ::-1, :]
io.imshow(I_rotated)
plt.axis("off")
```
