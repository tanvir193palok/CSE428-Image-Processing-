# Task 1

```python
# Import the "good.jpeg" image
I = io.imread('/content/drive/MyDrive/CSE428/LAB2/good.jpeg')

# normalize the image
I_norm = img_as_float(I) 

# perform arithmetic operations (contrast increase/decrease) on the image
I_add = np.clip(I_norm+0.5 ,0,1)# increase by 0.5
I_sub = np.clip(I_norm-0.4,0,1)# decrease by 0.4
I_mul = np.clip(I_norm*2,0,1)# multiply by 2
I_div = np.clip(I_norm/3,0,1) # divide by 3

fig, axes = plt.subplots(1,5,figsize = (30,30))
axes[0].imshow(I_norm)
axes[0].axis("off")
axes[0].set_title("Normal Image")

axes[1].imshow(I_add)
axes[1].axis("off")
axes[1].set_title("Increased by 0.5")

axes[2].imshow(I_sub)
axes[2].axis("off")
axes[2].set_title("Decreased by 0.4")

axes[3].imshow(I_mul)
axes[3].axis("off")
axes[3].set_title("Multiplied by 2")

axes[4].imshow(I_div)
axes[4].axis("off")
axes[4].set_title("Divided by 3")

plt.show()
```

```python
# Import the "land.jpeg" image
I = io.imread("/content/drive/MyDrive/CSE428/LAB2/good.jpeg")
I = color.rgb2gray(I)
# perform contrast stretching on the image
I_norm =np.clip(img_as_float(I),0,1)

hist , nbins = exposure.histogram(I_norm,nbins = 256,normalize = True)
plt.plot(nbins[1:],hist[1:])
```

```python
# perform contrast stretching on the image
I_stretched =np.clip(6.67*(I_norm-0.7),0,1)

hist , nbins = exposure.histogram(I_stretched,nbins = 256,normalize = True)
plt.plot(nbins[1:],hist[1:])
```
